```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data
df = pd.read_csv('/Users/vickancer/Downloads/movies.csv')
```

```python
df.head()
```

```python
# Lets see for any missing data's
# Finding NULL values in the dataset

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# DATA TYPES for the columns
df.dtypes
```

```python
# ecided to drop all rows that have missing data.

df = df.dropna()
```

```python
# Data type changed to int to remove '0' from the ends

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int')
```

```python
# create correct year column from the released column....just extracting YYYY

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# MY CORRELATION HYPOTHESIS:::
# 1. Budget high correlation
# 2. COmpany high correlation
```

```python
# BUILD Scatter plot BUDGET vs GROSS REVENUES

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Films')

plt.show()



```

```python
# Using seaborn for the REGRESSION PLOT ( budget vs gross)

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"teal"},line_kws={"color":"red"})
```

```python
# Let's start looking at the Correlation-- It will only work on Numerical Fields
# 3 types-- pearson(DEFAULT) , kendall , spearman

# WE WILL USE PEARSON

df.corr(method='pearson')  #Pearson
```

```python
 # HIGH correlation between budgets and gross
 # I was right
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
## NEXT --> Look at COMPANY
# Company is not Numeric, we have to convert it into that

df.head()
```

```python
## NEXT --> Look at COMPANY
# Company is not Numeric, we have to convert it into that

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
